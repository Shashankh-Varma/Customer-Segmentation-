```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv('/content/Mall_Customers.csv')
```

```python
customer_data.isnull().sum()
# TO find any missing values
```

```python
X = customer_data.iloc[:,[3,4]].values
# selecting Columns of annual income and spending score
```

```python
print(X)
```

```python
wcss = []

for i in range (1,11):

  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
#finding different number of clusters
# within culster sum of square
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('No Of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
Y = kmeans.fit_predict(X)
print(Y)


# return label for each data point based on their cluster
# optimal cluster is 5
# training the k means cluster
```

```python
plt.figure(figsize=(8,8))
#plotting culsters and their centroids
plt.scatter(X[Y==0,0],X[Y==0,1],s=50, c = 'green', label='Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50, c = 'red', label='Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50, c = 'yellow', label='Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50, c = 'orange', label='Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50, c = 'blue', label='Cluster 5')
#plotting the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 100,c='cyan',label='centroid')
plt.title('Customer Group')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()


#Visualizing all the cluster (which are 5 (0,1,2,3,4))
```
